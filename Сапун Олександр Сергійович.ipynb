```python
import numpy as np
np.random.seed(42)
print('NumPy version:', np.__version__)
```

## 1) Одновимірний масив (вектор) з першими 20-ма натуральними числами

```python
a = np.arange(1, 21)
print('a =', a)
print('shape:', a.shape)
```

## 2) Двовимірний масив (матриця) 4×4

```python
m_ones = np.ones((4, 4), dtype=int)
print('m_ones =\n', m_ones)
print('shape:', m_ones.shape)
```

## 3) Масив 5×5 і [-1; 1]

```python
m55 = np.random.uniform(-1, 1, size=(5, 5))
print('m55 =\n', m55)
```

## 4) Масив 3×3  [0; 1]

```python
m33_01 = np.random.random(size=(3, 3))
print('m33_01 =\n', m33_01)
```

## 5) Два одновимірні масиви розміром 6 (цілі числа 1..20) 

```python
x = np.random.randint(1, 21, size=6)
y = np.random.randint(1, 21, size=6)

print('x =', x)
print('y =', y)
print('\nПоелементне додавання x + y =', x + y)
print('Поелементне віднімання x - y =', x - y)
print('Поелементне множення x * y =', x * y)
```

## 6) Матриці 3×2 та 2×4 (цілі 1..5) 

```python
A = np.random.randint(1, 6, size=(3, 2))
B = np.random.randint(1, 6, size=(2, 4))

print('A =\n', A)
print('B =\n', B)
C = A @ B  
print('\nC = A @ B =\n', C)
print('shape C:', C.shape)
```

## 7) Матриця 3×3 (дійсні 0..1) 

```python
M = np.random.uniform(0, 1, size=(3, 3))
print('M =\n', M)
MT = M.T
print('\nM.T =\n', MT)
```

## 8) Дві матриці 3×3 (цілі 1..10) 

```python
P = np.random.randint(1, 11, size=(3, 3))
Q = np.random.randint(1, 11, size=(3, 3))
print('P =\n', P)
print('Q =\n', Q)
D = P - Q
print('\nD = P - Q =\n', D)
```

## 9) Дві матриці 3×3 (цілі -5..5) 

```python
R = np.random.randint(-5, 6, size=(3, 3))
S = np.random.randint(-5, 6, size=(3, 3))
print('R =\n', R)
print('S =\n', S)
Prod = R @ S
print('\nProd = R @ S =\n', Prod)
```

## 10) Матриця 3×3 (дійсні -1..1) 

```python
T = np.random.uniform(-1, 1, size=(3, 3))
print('T =\n', T)
row_sums = T.sum(axis=1).reshape(-1, 1)  
print('\nrow_sums (вектор-стовпчик) =\n', row_sums)
print('shape:', row_sums.shape)
```
